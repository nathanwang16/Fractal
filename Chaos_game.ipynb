```python
import numpy as np
import matplotlib.pyplot as plt

# Vertices of a triangle
vertices = np.array([
    [0, 0],
    [1, 0],
    [0.5, np.sqrt(3)/2]
])


# Start point inside the triangle
point = np.array([0, 0])

points = [point]

# Chaos Game: move halfway toward a randomly chosen vertex
for i in range(10000):
    chosen_vertex = vertices[np.random.choice([0, 1, 2])]
    print(f"Chosen vertex: {chosen_vertex}")
    point = (point + chosen_vertex) / 2
    points.append(point)
    print(point)
    
    i = i+1

# Plotting
points = np.array(points)
plt.figure(figsize=(6, 6))
plt.plot(points[:, 0], points[:, 1], 'k.', markersize=0.5)
plt.axis('equal')
plt.title("Chaos Game – Sierpinski Triangle")
plt.show()
```
